#Bag of Words Architecture:
Tokenize - Create Vocab - Count Frequency of Tokens - Create Vectorized Representation

```python
import pandas as pd
```

```python
from keras.preprocessing.text import Tokenizer
```

```python
text = [
        'A rose is a flower',
        'A fern is not a flower',
        'the dog ate the rose'
]
```

Create a pandas series object from the list of sentences

```python
len(text)
```

#[Keras Tokenizer](https://www.tensorflow.org/api_docs/python/tf/keras/preprocessing/text/Tokenizer)
tf.keras.preprocessing.text.Tokenizer<br>
method fit_on_texts

```python
model = Tokenizer()
model.fit_on_texts(text)
```

Double-click (or enter) to edit

```python
print(f'Key : {list(model.word_index.keys())}')
```

```python
len(list(model.word_index))
```

<font color='#4cd137'> text_to_matrix <br> numpy array with the shape len sequences * size vocab

```python
vec_rep = model.texts_to_matrix(text, mode='count')
vec_rep
```

The length of the vector will always be equal to<br>vocabulary size * len document
