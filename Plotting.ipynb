### PLOTTING

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("train.csv")
df.head()
```

```python
df.LotFrontage.plot(figsize=(5,5))
```

```python
df.LotFrontage.hist(bins=100) # bins is the level of detail
```

```python
df[['LotFrontage', 'LotArea', 'SalePrice']].plot()
```

### Boxplot

```python
df[['SalePrice']].boxplot()

# green line = the mean
# square = standard deviation
# horizontal lines = 25% and 75% quartile
# circles = plotted values
```

```python
df[['MSZoning', 'SalePrice']].boxplot(by='MSZoning')
```
