#Speech / Syntax / Parser

# nouns.
# constituents - group of words that behave like a single unit or phrase.
# determiner or DET - a word, phrase, or affix that occurs together with a noun or noun phrase.

# you need to find the constituents in the sentence.
# CFG (Context Free Grammar).
# Noun Phrase (NP) = DET + Nominal. ("the race car").
# Verb Phrase (VP) = Verb + range of combos (Verb+NP,Verb+PP,Verb+NP+PP). ("eating the pancakes")
# Prepositional Phrase (PP + NP). ("from the restaurant")

#S=[NP]+[VP]+[PP]

#NP        VP    PP
#The boy / ate the pancakes / from the restaurant

#P    NP
#under / a table

#Part of Speech Tagging

#NLTK Taggers
#This package contains classes for part-of-speech tagging, simply "tagging". A "tag" specifies some property of a token, such as its part of speech.
#An Off-the-shelf tagger is available for English.

#Functions
#punkt is used for tokenising sentences.
#averaged_perceptron_tagger is used for tagging words with their part of speech (POS).


#POS:

#NN = Noun, singular
#NNS = Noun, plural
#NNP = Proper noun, singular
#NNPS = Proper noun, plural
#EX = Existential there, such as in the sentence "There was a party"
#PRP = Personal pronoun (PP)
#PRP$ = Possessive pronoun (PP$)
#POS = Possessive ending
#RBS = Adverb, superlative
#RBR = Adverb, comparative
#RB = Adverb
#JJS = Adjective, superlative
#JJR = Adjective, comparative
#JJ = Adjective
#MD = Modal
#VB = Verb, base form
#VBP = Verb, present tense, other than 3rd person singular
#VBZ = Verb, present tense, 3rd person singular
#VBD = Verb, past tense
#VBN = Verb, past participle
#VBG = Verb, gerund or present participle

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
sentence = "The boy ate the pancakes from the restaurant"
```

Tokenize

```python
token = nltk.word_tokenize(sentence)
print(token)
```

POS Tagging

```python
tags = nltk.pos_tag(token)
print(tags)
```
