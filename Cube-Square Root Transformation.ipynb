# Cube / Square Root Transformation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('bigmart.csv')
```

```python
data.head()
```

## QQ Plot

A Q-Q plot is a scatterplot created by plotting two sets of quantities against one another. If both sets of quantities came from the same distribution, we should see the points forming a line that's roughly straight.

```python
# make the qq plot for the Item Outlet Sales to check for its distribution and skewness

from scipy import stats
from scipy.stats import norm

sns.distplot(data['Item_Outlet_Sales'], fit = norm)

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(data['Item_Outlet_Sales'])
print('mean = {:.2f} and standard deviation = {:.2f}'.format(mu, sigma))

# Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
          loc='best')
plt.ylabel('Frequence')
plt.title('Item Outlet Sales Distribution')

# Also get the QQ-plot
fig = plt.figure()
res = stats.probplot(data['Item_Outlet_Sales'], plot = plt)
plt.show()
```

```python
#Using the square root transformation

data['Item_Outlet_Sales'] = np.sqrt(data['Item_Outlet_Sales'])
```

```python
# After the transformation again checking for the distributions and the skewness

sns.distplot(data['Item_Outlet_Sales'], fit = norm)

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(data['Item_Outlet_Sales'])
print('mean = {:.2f} and standard deviation = {:.2f}'.format(mu, sigma))

# Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
          loc='best')
plt.ylabel('Frequence')
plt.title('Item Outlet Sales Distribution')

# Also get the QQ-plot
fig = plt.figure()
res = stats.probplot(data['Item_Outlet_Sales'], plot = plt)
plt.show()
```
