# NLP - TEXT CLASSIFICATION

* Text classification or Text categorization is the process of analyzing the natural language text and labelling the text with a predefined set of labels or tags.

* Text classifiers have proven to be a great alternative to structure textual data in a fast, cost-effective and scalable way.

* It allows to easily get insights from data and automate business processes.


Examples:
* Classifying emails as spam or not spam.
* Sentiment analysis: Understanding if the text has positive, negative or neutral sentiment.
* Language detection: Detecting the language of a given text.
* Classifying content into categories to easily search and navigate within a website or application.

Text classification is a very powerful tool for extracting value from unstructured data.

* We use supervised learning algorithms for classifying text.


# Types of applications for Text Classification

* Tagging content or products using categories as a way to improve browsing or to identify related content on the website.
    * E-commerce, news agencies, etc.

* As marketing is becoming more targeted every day, automated classification of users into cohorts can make a marketer's life simple.

* Text classification of content on websites help Google crawl websites easily which helps with SEO.

* Email providers use text classification to differentiate between legitimate and spam emails.


# Text Classification using Machine Learning

* Machine learning helps in text classification and classify based on past observations.

* By using pre-labelled examples as training data, ML algorithms learn different associations between pieces of text, and that a particular output is expected for a particular input.

* Machine learning algorithms work only with numerical data.

* Transform each piece of text into a numerical representation in the form of a vector.

# Naive Bayes Classifier and Algorithm

Naive Bayes is based on Bayes Theorem

### Conditional Probability

* Probability of an event occurring based on the occurrence of the previous event.

* We have a bag of 5 balls (2 red, 3 blue)
* We're asked to randomly take out a ball.
* What is the probability for red (2/5)
* What is the probability for blue (3/5)
* After taking out a red and blue ball it is (2/4) for both.

* The probability of 1 event depending on the probability of the previous event.

P(A/B) = P(A and B) / P(B) => P(A and B) = P(A/B) P(B)


# Bayes Theorem

* The probability of event A, given B == A and B, divided by probability of event B.

* Probability B, A===B and A, divided by probability of event A.


# Naive Bayes Classifier

* Naive Bayes is a family of probabilistics algorithm which uses Bayes theorem to predict the tag of the next.

* It is a probabilistic algorithm which means it calculates the probability of each tag for a given text.


### Example

We have 4 major words in the categories Sport and not sport.
    * 4 major words are, match, game, win and election
    
Sport                    Not sport

Match    6  (6/15)       Match    1  (1/15)
Game     5  (5/15)       Game     2  (2/15)
Win      3  (3/15)       Win      5  (5/15)
Election 1  (1/15)       Election 7  (7/15)


##### Categorize message with "win election" in it:

* the probability of a new message in "sport" or "not sport" category is 50%.

* P(sport/win election) = 0.006
* P(not sport/win election) = 0.07

* New message "win election" is in the "Not sport" category.


# Support Vector Machines
