```python
def removeCommas(text):
    data = []
    for i in text:
        if i == ',':
            data.append(' ')
        else:
            data.append(i)
    return data
```

```python
def addCommas(text):
    data = []
    for line in text:
        line[len(line) - 1] = str(',')
        data.append(line)
    return data

df['newl'] = addCommas(df['lines'])
```
