### Deleting a column

```python
df = pd.DataFrame(data = {"a": np.arange(1,10), "b": np.arange(10,19), "c": np.arange(20,29)})

df.drop('c', axis=1, inplace=True) # axis=1 means columns, axis=0 means rows 
df
```

### Deleting a row

```python
df = pd.DataFrame(data = {"a": np.arange(1,10), "b": np.arange(10,19), "c": np.arange(20,29)})

df.drop(2, axis=0, inplace=True)
df
```

### Deleting based on condition

```python
df = pd.DataFrame(data = {"a": np.arange(1,10), "b": np.arange(10,19), "c": np.arange(20,29)})
df.drop(df[df.a < 3].index) # because

df[df.a <3] # these got dropped
```

### Renaming columns

```python
df = pd.DataFrame(data = {"a": np.arange(1,10), "b": np.arange(10,19), "c": np.arange(20,29)})
df.rename(columns={'a': 'New1', 'b': 'Newb'}, inplace=True) # returns a copy so don't forget inplace=True
df
```
