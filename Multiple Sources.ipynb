```python
import numpy as np
import pandas as pd

df1 = pd.DataFrame(data = {"a": [1,2,1,2,1,2], "b": [3,4,3,4,3,4]})
df2 = pd.DataFrame(data = {"a": [1,3,1,3,1], "b": [5,4,5,4,5]})
```

```python
temp = pd.concat([df1, df2, df1, df2]) # must be in array
temp
```

```python
temp.index
```

```python
temp.loc[0, :] # loc refers to the value of index/column
```

```python
temp.iloc[0, :] # iloc refers to the index of index/column

# loc is the value selector ;; iloc is the index selector
```

```python
temp.reset_index()
```

```python
temp.reset_index(drop=True)
```

```python
# right concat
temp = pd.concat([df1, df2], axis=1)
temp
```

### Joining

```python
df1 = pd.DataFrame({
    "first": [1,1,1,1],
    "second": [1,1,1,1],
    "third": [1,2,1,2],
}, index=[1,2,3,4])

df2 = pd.DataFrame({
    "first": [2,2,2,2],
    "second": [2,2,2,2],
    "fourth": [1,3,1,3],
}, index=[3,4,5,6])
```

```python
df1
```

```python
df2
```

```python
# joining on columns
pd.concat([df1, df2], join='outer')
```

```python
# joining on columns
pd.concat([df1, df2], join='outer', axis=1)
```

```python
# joining on columns
pd.concat([df1, df2], join='inner', axis=1)
```

```python
# left join using join method. Concat can join x dataframes at once
df1.join(df2, how="left", rsuffix="_second")
```
