### Building Convolutional Neural Network

The building blocks of a convolution network are:

1. The convolutional layer.
2. Pooling.
3. Final layers.


### Convolutional layer

1. This layer consists of padding, convolution operation and Relu.

2. This layer receives the image and applies padding on top.

3. Select the dimensions of the filter and number of filters to apply convolution operation.

4. Start with a filter size of 7x7 or 5x5 and the number of filters can be around 16 or 32.

5. The output feature map after convolution is passed into the activation function such as Relu which will remove any negative signals.


### Pooling layer

1. Takes in the feature map as input and reduces its size by selecting Max or Average over its size.
    a. Pooling is generally applied over a 2x2 region.
    
2. Repeat these convolution and pooling layers any number of times.
    a. Increasing parameters to learn thus increasing the size of the model.

3. The input of the 2nd pooling layer would be the output of the 1st pooling layer.

### Final layers

* Final feature is extracted in the 2nd layer.

1. Assume the dimension of the feature map is 6x6x128 where 6x6 is the size of the feature, and 128 is the number of filters in the 2nd convolution layer.

2. We cannot map a 3 dimension output directly to the output layer, so we can either flatten the 3d vector and apply Global pooling.

### Method

1. Applying Global max pooling on a 6x6x128 returns an output of 1x1x128 which can be treated as 128 neurons.

2. Add in a few Dense layers which are used to predict the output.
    a. Use Softmax or Sigmoid in the final layer.
