#Tokenization Methods Using NLTK

Tokenization is making the structure of words (tokens) machine readable. This process of which tokenization is part is also called "preprocessing data".

hellohowareyoutoday = hello/how/are/you/today <-- (tokenized).

#Cliticization:

He's <-- the clitic is 's.
He is --> the cliticized word expands into "he is".

The tokenizer or the contractions library can be used to expand clitics.

```python
import nltk #importing nltk package
```

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
article = '''Did you know that World War 1 brought German Shephers to the Western Hemisphere? Rin Tin Tin was a German Shepherd rescured from WWI. The are some of the smartest dogs!'''
```

```python
from nltk.tokenize import sent_tokenize
```

```python
import contractions
```

```python
contracted_text = '''She's going for a walk. Jack didn't go because he was tired.
          They'd definitely go tonight. I'm sorry, I'll be late.
          That's the way I like 'em!! '''
```

```python
expanded_all = []
for word in contracted_text.split():
  expanded_all.append(contractions.fix(word))

expanded_clitics = ' '.join(expanded_all)
print('Text with contractions: ' + contracted_text)
print('Expanded Clitics: ' + expanded_clitics)
```

```python
word_tokenize(contracted_text)
```

```python
word_tokenize(expanded_clitics)
```
