# LOGISTIC REGRESSION & SIGMOID THEORY


#### Logistic Regression

* not a regression model but it's actually a classification model.
* it's a supervise machine learning algorithm.
* We cannot use mean squared error in case of logistic regression because mean squared error is not a convex function when used in logistic regression. Hence, we can be stuck in a local minima instead of reaching global minima.

- https://towardsdatascience.com/why-not-mse-as-a-loss-function-for-logistic-regression-589816b5e03c

* We prefer a model with minimum AIC value.

* Linear regression error values have to be normally distributed but in the case of logistic regression this is not the case.

#### Thresholding

1. Converts the sigmoid outputs into predictions.

2. Involves setting up a statistical threshold.
    a. Any value equal or above the threshold is given the upper limit and value below the threshold will be given the lower limit.
    
If Threshold = 0.5

a. Sigmoid(Z) = 0.49 then Prediction = 0
b. Sigmoid(Z) = 0.51 then Prediction = 1
