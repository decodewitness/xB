```python
import twint
import pandas as pd
import nest_asyncio
nest_asyncio.apply()
```

```python
c = twint.Config()
c.Pandas = True
c.Store_csv = True
c.Search = "bitcoin"
c.Limit = 5000
c.Since = "2021-01-01"
c.Until = "2021-04-30"
c.Lang = 'en'
```

```python
twint.run.Search(c)
```

```python
twint.storage.panda.Tweets_df.head()
```

#Configure To Scrape Based On Username

```python
c = twint.Config()
c.Pandas = True
c.Username = "elonmusk"
c.Limit = 5000
c.Since = "2022-01-01"
c.Until = "2022-05-20"
c.Store_csv = True
```

```python
twint.run.Search(c)
```

```python
twint.storage.panda.Tweets_df['tweet'].count()
```

```python
twint.storage.panda.Tweets_df['tweet'][122]
```

<font color='aubergine'>Scraping Geo Tag Data

```python
c = twint.Config()
c.Pandas = True
c.Geo = "51.5, -0.127, 15km"
c.Limit = 5000
c.Search = "bitcoin"
c.Since = "2016-01-01"
c.Until = "2026-05-20"
c.Store_csv = True
```

```python
twint.run.Search(c)
```

```python
twint.storage.panda.Tweets_df['tweet'].count()
```
