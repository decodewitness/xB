# Convolutional Operation & Filters in CNN

* Filters also called Kernels, are used to enhance the features of the image by element wise multiplying a matrix over the whole image.

    * It's a matrix, multiplied over the whole image and extracting features.
    * It is called Convolution Operation.

# How Convolution works?

1. Place the filter on top of the image and perform an element wise product which results in 9 numbers.

2. Add up all the numbers which will be placed on the resultant matrix which is the output.

3. Repeat the process by moving the filter over the entire image from top left to bottom right and fill up the resultant matrix which is our final feature matrix.

4. If the filter learns to identify the pattern of the object, it would be able to identify that object regardless of its position in the image. Hence overcomes Translation Invariance.


# Need for Multiple Filters

1. One filter when convoluted on the image would return in one feature map.
    
    a. To extract multiple features from an image we need to use multiple filters.
    

# Filter for RGB Images

* The images can be 3 dimensional as well, for example a color image which has RGB filters as 3rd dimension.

* In this case our filter should also be 3 dimensional to be able to perform an element wise product.


# Filters as Weights of the Network

1. Convolutional neural networks solve this by considering the filter to be a matrix of learnable weights which will be updated and perform backpropagation.

2. The feature generated by convoluting images and filters are passed into the activation function (like Relu or Leaky Relu).
