#Grammar Rules

Context-Free

**S -> NP VP** 

A sentence could be followed by a **noun phrase** and a **verb phrase**:

"We | want **a morning flight**"

-----> **V** ("want") ---> **NP** ("a morning flight")

**NP** ("We") | (**V** "want") + (**NP** "a morning flight") = **(NP) | (VP)**

Rule: **(S=Sentence -> Noun Phrase, Verb Phrase)**

#Examples

We | want a morning flight. (V NP PP) from Manchester (PP)

NP --> {We}
Proper Noun --> {Manchester}
VP --> {want a morning flight}

I | want a flight (VP) --> a (DET) flight (N)

DET Nominal --> {a flight}

Nominal:
 -- Nominal Noun --> {morning + flight}
 -- Noun --> {flights}

 I want a flight from Manchester [NP][VP]->[V][NP][PP]

 PP (P NP) --> {from + Manchester (NP)}
 
VP (Verb) --> {do} 
(V + NP) --> {want a flight}
V + NP + PP --> {leave + Manchester + in the morning}
V + PP --> {on Friday} (PP + Noun) = (PP)


He(NP) | wants(V) a flight | on Friday(PP)
