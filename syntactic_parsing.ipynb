#Syntactic Parsing

The parser will figure out which words belong together (as "phrases"), and which words are the subject or object of a verb.
(The Stanford Parser, developed by the Stanford NLP Group - was one of the biggest breakthroughs in natural language processing.)

What songs were released by British music bands in 2020?

main noun = songs.
subject = what songs.
noun phrase = what songs.
